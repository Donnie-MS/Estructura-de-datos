```
#N y M positivos
#NxM = N + N*(M-1)   si M distinto de 1
#      N             si M igual 1
def producto(N:int, M:int)->int:
  salida = None
  if M == 1:
    salida = N
  else:
    salida = N + producto(N,M-1)
  return salida
```

### **Ejercicio 1**

Implementar una función que calcule el factorial de un número de forma recursiva

```
def factorial(N:int) -> int:
  salida = None
  if N <= 0:
      salida = 1
  else:
      salida = N * factorial(N-1)
  return salida

print(factorial(3))
```

```
def cantidadTotalBaldosas(dias:int, dia:int=1, anterior:int=100, previo:int=200)->int:
    if dia > dias:
        return 0
    if dia == 1:
        cantidad = 100
    elif dia % 2 == 0:
        cantidad = anterior * 2
    else:
        cantidad = anterior + previo
    return cantidad + cantidadTotalBaldosas(dias, dia + 1, cantidad, anterior)

dias = int(input("Introduce la cantidad de días: "))
totalBaldosas = cantidadTotalBaldosas(dias)
print(f"En {dias} días se colocaron un total de {totalBaldosas} baldosas.")
```

```
def cantBaldosas(dia:int)->int:
  salida = None
  if dia == 1:
    salida=100
  elif dia > 1:
    if dia%2 == 0:
      salida = cantBaldosas(dia-1)*2
    else:
      salida = cantBaldosas(dia-2)+cantBaldosas(dia-1)
  return salida

print(cantBaldosas(500))
```

```
def cantBaldosasAcumuladas(dias:int)->int:
  baldosasTotales = 0
  for dia in range(1,dias+1):
    baldosasTotales += cantBaldosas(dia)
  return baldosasTotales
```

### **Ejercicio 16**
Implementar una función recursiva para encontrar el máximo elemento de un arreglo

```
import numpy as np
def maximo(v:np.array(int))->int:
  valorMaximo = None
  if len(v) == 1:
    valorMaximo = v[0]
  else:
    valorMaximo = max(v[0], maximo(v[1:]))
  print(v, f"maximo={valorMaximo}")
  return valorMaximo

print(maximo(np.array([2,3,10,3])))
```

```
import numpy as np
def maximo(v:np.array(int), posInicio:int=0)->int:
  valorMaximo = None
  if posInicio == len(v) - 1:
    valorMaximo = v[posInicio]
  else:
    valorMaximo = max(v[posInicio], maximo(v, posInicio+1))
  print(v, f"pos={posInicio}", f"maximo={valorMaximo}")
  return valorMaximo

print(maximo(np.array([2,3,10,3])))
```
